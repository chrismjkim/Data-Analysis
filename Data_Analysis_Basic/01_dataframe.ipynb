### **(1) 데이터프레임 기본 조회**

```python
import pandas as pd
```

2) read_csv()로 DataFrame 생성

- 파일 경로: https://raw.githubusercontent.com/DA4BAM/dataset/master/titanic_simple.csv

**[titanic_simple 데이터 셋 정보]**

- PassengerId : 승객번호
- Survived : 생존여부(1:생존, 0:사망)
- Pclass : 객실등급(1:1등급, 2:2등급, 3:3등급)
- Name : 승객이름
- Sex : 성별(male, female)
- Age : 나이
- Fare : 운임($)
- Embarked : 승선지역(Southhampton, Cherbourg, Queenstown)

```python
file_dir = "https://raw.githubusercontent.com/DA4BAM/dataset/master/titanic_simple.csv"
titanic = pd.read_csv(file_dir)
```

6) info(): 열 값, 개수, 데이터타입 확인

```python
# 열 정보 확인
print(titanic.info())
```

7) value_counts(): 열 별 고유값, 개수

### **(2) 데이터프레임 조건 조회**

1. 특정 열 조회

```python
titanic[["Name"]]
```

2. 복수 열 조회

```python
titanic[["Name", "Age", "Sex"]]
```

3. 복수 열의 맨 위 행들 조회

```python
titanic[["Name", "Age", "Sex"]].head(10)
```

4. 나이가 70세 이상인 행 확인

```python
is_older_than_70 = titanic["Age"] >= 70
titanic[is_older_than_70]
```

5. Fare 평균

```python
Fare_mean = titanic["Fare"].mean()
print(f"Fare(운임)의 평균: {Fare_mean}")
```

6. Fare의 평균보다 값이 작은 행 조회

```python
has_paid_less = titanic["Fare"] < Fare_mean
print(titanic[has_paid_less])
```

7. 승선지역(Embarked)이 'Southhampton', 'Queenstown' 인 승객

```python
# 데이터 확인 결과, Southhampton에서 승선한 승객은 없음
em_Sh = titanic["Embarked"] == 'Southhampton'
em_Qt = titanic["Embarked"] == 'Queenstown'
print(titanic[em_Sh | em_Qt])
```

8. 나이(Age)가 10대(10이상, 20 미만)인 남자 청소년의 평균 운임(Fare)

```python
is_teen1 = titanic["Age"] >= 10
is_teen2 = titanic["Age"] < 20
print(titanic[is_teen1 & is_teen2])
```
