# **시각화 기초**

## **1.환경준비**

### **(1) 라이브러리**

```python
import pandas as pd
import numpy as np

# 시각화 라이브러리
import matplotlib.pyplot as plt # 그래프를 그리는 함수들을 담은 모듈
import seaborn as sns
```

### **(2) 데이터 불러오기**

**[airquality 데이터 셋 정보]**

- Ozone: 오존 농도  
- Solar.R: 태양복사량
- Wind: 풍속
- Temp: 기온
- Date : 연,월,일

```python
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/airquality_simple2.csv'
data = pd.read_csv(path)
data['Date'] = pd.to_datetime(data['Date'])
data.dropna(axis = 0, inplace = True)
data.head()
```

## **2.기본차트 그리기**

### **(1) 차트 그리기**

* plt.plot(1차원 값)
    * x축 : 인덱스
    * y축 : 1차원 값
    * 라인차트 생성

```python
plt.plot(data['Temp']) # Series 반환
plt.show()
```

### **(2) x축과 y축의 값 지정**
* plt.plot(x축, y축)

```python
plt.plot(data['Date'], data['Temp']) # x축에는 날짜, y축에는 온도
plt.show()
```

* plt.plot(x축, y축, data = )


* x 축 : Date
* y 축 : Ozone


```python
plt.plot('Date', 'Ozone', data = data)
```

## **3.차트 꾸미기**

https://matplotlib.org/2.1.2/api/_as_gen/matplotlib.pyplot.plot.html

### **(1) x축, y축 이름, 타이틀 붙이기**

```python
plt.plot(data['Date'], data['Ozone'])

plt.xticks(rotation = 30)       # x축 값 꾸미기 : 30도 회전
plt.xlabel('Date')              # x축 이름 지정
plt.ylabel('Ozone')             # y축 이름 지정
plt.title('Daily Airquality')   # 타이틀

plt.show()
```

### **(2) 라인 스타일 조정**

* color=  
    * 'red','green','blue' ...
    * 혹은 'r', 'g', 'b', ...
    * https://matplotlib.org/stable/gallery/color/named_colors.html
* linestyle=
    * 'solid', 'dashed', 'dashdot', 'dotted'
    * 혹은 '-' , '--' , '-.' , ':'
* marker=

| marker |	description |
| ---- | ---- |
|"." |	point |
|"," |	pixel |
|"o" |	circle |
|"v" |	triangle_down |
|"^" |	triangle_up |
|"<" |	triangle_left |
|">" |	triangle_right |

```python
plt.plot(data['Date'], data['Ozone']
        ,color='green'                # 컬러
        , linestyle='dotted'          # 라인스타일
        , marker='o')                 # 값 마커(모양)

plt.xlabel('Date') # x축의 이름
plt.ylabel('Ozone') # y축의 이름
plt.title('Daily Airquality') # 그래프의 제목
plt.xticks(rotation=45) # x축 틱의 레이블 회전

plt.show()
```

### **(3) 여러 그래프 겹쳐서 플롯**

```python
# 첫번째 그래프
plt.plot(data['Date'], data['Ozone'], color='green', linestyle='dotted', marker='o')
# 두번째 그래프
plt.plot(data['Date'], data['Temp'], color='r', linestyle='-', marker='s')

plt.xlabel('Date')
plt.ylabel('Ozone')
plt.title('Daily Airquality')
plt.xticks(rotation=45)

plt.show()
```

### **(4) 범례, 그리드 추가**

```python
plt.plot(data['Date'], data['Ozone'], label = 'Ozone')  # label = : 범례추가를 위한 레이블값
plt.plot(data['Date'], data['Temp'], label = 'Temp')

plt.legend(loc = 'upper right')    # 범례 표시하기. loc = : 위치
plt.grid()

plt.show()
```

그래프 연습


```python
plt.plot(data['Date'], data['Ozone'], label = 'Ozone')
plt.plot(data['Date'], data['Temp'], label = 'Temperature')
plt.plot(data['Date'], data['Wind'], label = 'Wind')
plt.xlabel('Date')
plt.xticks(rotation = 30)
plt.title('Airquality')
plt.legend(loc = 'upper right')
plt.grid()
plt.show()
```

## **4.추가기능**

### **(1) 데이터프레임.plot()**
* https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html

```python
data.plot(x = 'Date', y = ['Temp','Ozone']
        , title = 'Daily Airquality')
plt.grid()
plt.show()
```

```python
data.plot(x= 'Date', y = ['Ozone', 'Temp', 'Wind'], 
    title = 'Airquality', legend = True, grid = True, figsize = (12, 8))
# 참고: legend parameter의 default = True
```

### **(2) 축 범위 조정**

```python
plt.plot(data['Ozone'])

plt.ylim(0, 100) # y축 값 0부터 100까지
plt.xlim(0,10) # x축 값 0부터 10까지
plt.grid()
plt.show()
```

### **(3) 그래프 크기 조정**

* default size는 6.4, 4.4

```python
plt.figure(figsize = (4, 3))  # plot이 아닌 figure에서 편집

plt.plot(data['Ozone'])

plt.ylim(0, 100)
plt.xlim(0,10)

plt.grid()
plt.show()
```

### **(4) 수평선 수직선 추가**

```python
plt.plot(data['Ozone'])

plt.axhline(40, color = 'grey', linestyle = '--') # horizontal line
plt.axvline(10, color = 'red', linestyle = '--') # vertical line
plt.show()
```

### **(5) 그래프에 텍스트 추가**

```python
plt.plot(data['Ozone'])

plt.axhline(40, color = 'grey', linestyle = '--') 
plt.axvline(10, color = 'red', linestyle = '--') 

plt.text(5, 41, '40')
plt.text(10.1, 20, '10')

plt.show()
```

그래프 연습

```python
plt.plot(data['Date'], data['Ozone'])

ozone_mean = data['Ozone'].mean()
plt.xticks(rotation = 45)
plt.axhline(ozone_mean, color = 'red', linestyle = '--')
plt.text(pd.to_datetime('1973-06-01'), 41, round(ozone_mean, 2))
plt.show()
```

## **5.추가기능: 여러 그래프 나눠서 그리기**

![](https://codetorial.net/matplotlib/_images/subplot_03.png)

* row : 고정된 행 수
* column : 고정된 열 수
* index : 순서

### **(1) 위, 아래로(3행 1열) 그래프 그리기**

```python
data.head()
```

```python
plt.figure(figsize = (12,8))
plt.subplot(3,1,1) # 3행 1열의 첫 번째 그래프
plt.plot('Date', 'Temp', data = data)
plt.grid()

plt.subplot(3,1,2) # 3행 1열의 두 번째 그래프
plt.plot('Date', 'Wind', data = data)

plt.subplot(3,1,3) # 3행 1열의 세 번째 그래프
plt.plot('Date', 'Ozone', data = data)
plt.grid()
plt.ylabel('Ozone')

plt.tight_layout() # 그래프간 간격을 적절히 맞추기
plt.show()
```

### **(2) 옆으로 (1행 3열) 그래프 그리기**

```python
plt.figure(figsize = (15,5))
plt.subplot(1,3,1) # 1행 3열의 첫 번째 그래프
plt.plot('Date', 'Temp', data = data)
plt.title('Temp')
plt.xticks(rotation = 40)
plt.grid()

plt.subplot(1,3,2) # 1행 3열의 두 번째 그래프
plt.plot('Date', 'Wind', data = data)
plt.title('Wind')
plt.xticks(rotation = 40)
plt.grid()

plt.subplot(1,3,3) # 1행 3열의 세 번째 그래프
plt.plot('Date', 'Ozone', data = data)
plt.title('Ozone')
plt.xticks(rotation = 40)
plt.grid()

plt.tight_layout() # 그래프간 간격을 적절히 맞추기
plt.show()
```
