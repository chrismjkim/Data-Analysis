#### **2) 밀도함수 그래프(kde plot)**

* 히스토그램의 단점
    * 구간(bin)의 너비를 어떻게 잡는지에 따라 전혀 다른 모양이 될 수 있음
>
* 밀도함수 그래프
    * 막대의 너비를 가정하지 않고 모든 점에서 데이터의 밀도를 추정하는 커널 밀도 추정(Kernel Density Estimation)방식을 사용하여 이러한 단점을 해결.
    * 밀도함수 그래프 아래 면적은 1

* 밀도함수 그래프 그리기

```python
sns.kdeplot(x='Fare', data = titanic) # Kernel Density Estimation

plt.show()
```

* 밀도 추정
    * 측정된(관측된) 데이터로부터 전체 데이터 분포의 특성을 추정
    * 예를 들어… OO역 사거리 일일 교통량을 측정한다고 해 봅시다.
        * 어제는 1200대 차량이 통과했고, 오늘은 1420대, 내일은, 모레는…
        * 이렇게 3개월간 매일 측정했다고 할 때,
        * 우리는 약 90일치의 데이터를 가지고 일일 교통량 분포를 히스토그램으로 그려볼 수 있습니다.
        * 그리고 나서 특정한 날의 교통량이 얼마나 될지 확률로 나타냄.



titanic Age에 대해서 밀도함수

* 보스톤 집값 데이터를 이용

**변수설명**
* medv  : 1978 보스턴 주택 가격, 506개 타운의 주택 가격 중앙값 (단위 1,000 달러) <== Target
* crim	범죄율
* zn	25,000 평방피트를 초과 거주지역 비율
* indus	비소매상업지역 면적 비율
* chas	찰스강변 위치(범주 : 강변1, 아니면 0)
* nox	일산화질소 농도
* rm	주택당 방 수
* age	1940년 이전에 건축된 주택의 비율
* dis	직업센터의 거리
* rad	방사형 고속도로까지의 거리
* tax	재산세율
* ptratio	학생/교사 비율
* lstat	인구 중 하위 계층 비율


```python
# 보스톤 집값 데이터
boston = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/boston.csv')
boston.head()
```

## (2) 단변량 분석

* medv(집값)

```python
medv_stats = boston['medv'].describe()
print(medv_stats)
plt.boxplot(boston['medv'], vert=False)
plt.show()
print(f"사분위수: 25%: {medv_stats['25%']}, 50%: {medv_stats['50%']}, 75%: {medv_stats['75%']}")
```

```python
sns.histplot(x='medv', data=boston, kde=True)
print(f"평균: {medv_stats['mean']}, 표준편차: {medv_stats['std']}")
```

* crim(범죄율)

```python
crim_stats = boston['crim'].describe()
print(crim_stats)
plt.boxplot(boston['crim'], vert=False)
plt.show()
print(f"사분위수: 25%: {crim_stats['25%']}, 50%: {crim_stats['50%']}, 75%: {crim_stats['75%']}")

# 사분위수와 상자 그림으로는 쉽게 판단할 수 없고, 이상치가 매우 많은 것을 알 수 있다.
# 봉우리의 높이가 어느 정도인지 상자 그림으로 명확하게 판단할 수 없다.
```

```python
sns.histplot(x='crim', data=boston, bins = 25, kde=True)
print(f"평균: {crim_stats['mean']}, 표준편차: {crim_stats['std']}")

# 오른쪽으로 치우친 분포를 가지고 있음을 알 수 있다 (positively skewed).
```

* lstat(하위계층 비율)

```python
lstat_stats = boston['lstat'].describe()
print(lstat_stats)
plt.boxplot(boston['lstat'], vert=False)
plt.show()
print(f"사분위수: 25%: {lstat_stats['25%']}, 50%: {lstat_stats['50%']}, 75%: {lstat_stats['75%']}")

# 보스턴의 도시 중 50%는 저소득층 비율이 7% ~ 17% 사이에 위치한다.
```

```python
sns.histplot(x='lstat', data=boston, bins = 25, kde=True)
print(f"평균: {lstat_stats['mean']}, 표준편차: {lstat_stats['std']}")

# 오른쪽으로 치우친 분포를 가지고 있음을 알 수 있다 (positively skewed).
```

* ptratio(교사1명당 학생수)

```python
ptratio_stats = boston['ptratio'].describe()
print(ptratio_stats)
plt.boxplot(boston['ptratio'], vert=False)
plt.show()
print(f"사분위수: 25%: {ptratio_stats['25%']}, 50%: {ptratio_stats['50%']}, 75%: {ptratio_stats['75%']}")
```

```python
sns.histplot(x='ptratio', data=boston, bins = 25, kde=True)
print(f"평균: {ptratio_stats['mean']}, 표준편차: {ptratio_stats['std']}")

# 특정 값만 매우 높은 분포를 띄고 있음을 알 수 있다.
# 교육정책, 인사발령 등에 의해 조정되고 있을 가능성이 높음을 시사한다고 이야기할 수 있다.
```
